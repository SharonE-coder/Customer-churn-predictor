# Model Evaluation

```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1 Load model and data
model = joblib.load("C:/projects/Customer-churn-predictor/models/logistic_regression.pkl")
data = pd.read_csv("C:/projects/Customer-churn-predictor/data/processed/telco_churn_clean.csv")  # use lowercase 'processed'

# 2 Prepare features
X = data.drop("Churn", axis=1)

# 3 Get coefficients as feature importance
importances = np.abs(model.coef_[0])  # absolute value = importance magnitude

# 4 Create pandas Series
feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=True).tail(10)

# 5 Plot
plt.figure(figsize=(8,5))
feat_imp.plot(kind='barh', color='steelblue')
plt.title("Top 10 Most Important Features (Logistic Regression)")
plt.xlabel("Coefficient Magnitude (Importance)")
plt.tight_layout()
plt.show()

```
